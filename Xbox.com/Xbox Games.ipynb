```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import requests
import json
import pandas as pd
driver = webdriver.Chrome()
```

```python
driver.get("https://www.xbox.com/en-us/games/all-games?cat=all")
```

```python
big_ids = []
games = driver.find_elements_by_css_selector("div.m-product-placement-item.gameDiv")
for game in games:
    big_ids.append(game.get_attribute('data-bigid'))
    
print(len(big_ids))
```

```python
import pickle
with open('big_ids.pickle', 'wb') as handle:
    pickle.dump(big_ids, handle)
```

```python
big_ids_strings = []
last_index = 0
for j in range(1,2585):
    if j%20 == 0:
        joined_string = ",".join(big_ids[last_index:j])
        last_index = j
        print(j)
        big_ids_strings.append(joined_string)

joined_string = ",".join(big_ids[2580:])
big_ids_strings.append(joined_string)
```

```python
big_ids_strings[0]
```

```python
url = f"https://displaycatalog.mp.microsoft.com/v7.0/products?bigIds={big_ids_strings[0]}&market=US&languages=en-us&MS-CV=DGU1mcuYo0WMMp+F.1"

payload={}
headers = {
  'authority': 'displaycatalog.mp.microsoft.com',
  'sec-ch-ua': '"Chromium";v="92", " Not A;Brand";v="99", "Google Chrome";v="92"',
  'accept': '*/*',
  'dnt': '1',
  'sec-ch-ua-mobile': '?0',
  'user-agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.107 Safari/537.36',
  'origin': 'https://www.xbox.com',
  'sec-fetch-site': 'cross-site',
  'sec-fetch-mode': 'cors',
  'sec-fetch-dest': 'empty',
  'referer': 'https://www.xbox.com/',
  'accept-language': 'en-US,en;q=0.9,fr-FR;q=0.8,fr;q=0.7,ar;q=0.6'
}

response = requests.request("GET", url, headers=headers, data=payload)
```

```python
response.json()
```

```python
all_games = []
for p in response.json()['Products']:
    game = {
        "id" : p['ProductId'],
        "issue_date": p['MarketProperties'][0]['OriginalReleaseDate'],
        "min_age": p['MarketProperties'][0]['MinimumUserAge'],
        "category": p['Properties']['Category'],
        "developer": p['LocalizedProperties'][0]['DeveloperName'],
        "publisher":p['LocalizedProperties'][0]['PublisherName'],
        "decription": p['LocalizedProperties'][0]['ProductDescription'],
        "title": p['LocalizedProperties'][0]['ProductTitle'] + " Xbox",
        "image": p['LocalizedProperties'][0]['Images'][0]['Uri'],
    }
    all_games.append(game)
all_games
```

```python
df = pd.DataFrame(all_games)
df.to_csv('xbox_games.csv')
```
