```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import pickle
import json
```

```python
class ShopifyScraper():
    def __init__(self, website):
        self.website = website
        self.all_products = []
        
    def get_data(self):
        page = 1
        
        while True:
            r = requests.get(f'{self.website}/products.json?limit=250&page={page}')
            num_products = len(r.json()['products'])
            print(f"current page: {page}")
            print(num_products)
            
            for p in r.json()['products']:
                for v in p['variants']:
                    details = {
                        "id" : v['id'],
                        "sku" : v['sku'],
                        "name" : p['title'],
                        "handle" : p['handle'],
                        "type" : p['product_type'],
                        "price" : v['price'],
                        "image" : p['images'][0]['src']
                    }
                    self.all_products.append(details)
            if num_products < 250:
                break
            else:
                page = page + 1

        return self.all_products
```

```python
HelmBoots = ShopifyScraper("https://helmboots.com")
data = HelmBoots.get_data()
df = pd.DataFrame(data)
df['handle'].value_counts()
```
